#Найдите частные производные сложной функции E = (y^ - y)^2, y^=wx+b по переменным w и b

Для нахождения частных производных функции E = (y^ - y)^2 по переменным w и b, сначала найдем частную производную по переменной y^

Найдем частную производную функции E по переменной y^:

∂E/∂y^ = 2(y^ - y) * ∂(y^ - y)/∂y^ = 2(y^ - y) * 1 = 2(y^ - y)

Теперь найдем частную производную функции E по переменной w, используя найденную выше производную и правило дифференцирования сложной функции:

∂E/∂w = ∂E/∂y^ * ∂y^/∂w = 2(y^ - y) * ∂(wx + b)/∂w = 2(y^ - y) * x = 2x(y^ - y)

Наконец, найдем частную производную функции E по переменной b, используя ту же производную по y^:

∂E/∂b = ∂E/∂y^ * ∂y^/∂b = 2(y^ - y) * ∂(wx + b)/∂b = 2(y^ - y) = 2(y^ - y)

Таким образом, частные производные функции E = (y^ - y)^2 по переменным w и b равны соответственно 2x(y^ - y) и 2(y^ - y).


#Найдите частные производные сложной функции E = ylny^ + (1-y)ln(1-y^), y^ = 1/(1+e^(-z)) по переменной z

Найдем частную производную функции E по переменной y:

dE/dy = ln y^ + y * d(ln y^)/dy - ln(1-y^) - (1-y) * d(ln(1-y^))/dy

Здесь нам понадобится правило дифференцирования логарифма:

d(lnu)/du = 1/u

dE/dy = ln y^ + y * (1/y^) - ln(1-y^) - (1-y) * (-1/(1-y^)) = ln y^ + 1 - ln(1-y^) - 1/(1-y^)

Теперь найдем частную производную функции E по переменной z, используя найденную выше производную по y и цепное правило дифференцирования:

Для начала найдем частную производную y^ по переменной z:

d(y^)/dz = d(1/(1+e^(-z)))/dz = -(1/(1+e^(-z))^2) * d(1+e^(-z))/dz
= -(1/(1+e^(-z))^2) * (-e^(-z)) = e^(-z)/(1+e^(-z))^2

Теперь подставим частную производную y^ по z и частную производную E по y в цепное правило:

dE/dz = dE/dy * dy/dz = (ln y^ + 1 - ln(1-y^) - 1/(1-y^)) * e^(-z)/(1+e^(-z))^2

Таким образом, мы найдем частную производную функции E = ylny^ + (1-y)ln(1-y^) по переменной z.
